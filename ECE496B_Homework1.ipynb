```python
from datasets import load_dataset
from transformers import AutoTokenizer

# Load Hugging Face Tokenizers
tinystories_tokenizer = AutoTokenizer.from_pretrained("EleutherAI/gpt-neo-125M")  # Closest to TinyStories
owt_tokenizer = AutoTokenizer.from_pretrained("gpt2")  # GPT-2 tokenizer for OpenWebText

# Load datasets
tinystories = load_dataset("text", data_files="/content/data/TinyStoriesV2-GPT4-train.txt", split="train")
owt = load_dataset("text", data_files="/content/data/owt_train.txt", split="train")

import numpy as np

num_samples = 10

# Ensure dataset is accessible
tinystories_train = tinystories.shuffle(seed=42)
owt_train = owt.shuffle(seed=42)

# Sample 10 random documents
tinystories_docs = [tinystories_train[i]["text"] for i in range(num_samples)]
owt_docs = [owt_train[i]["text"] for i in range(num_samples)]

def tokenize_docs(tokenizer, docs):
    return [tokenizer.encode(doc) for doc in docs]

# Tokenize using respective tokenizers
tinystories_tokens = tokenize_docs(tinystories_tokenizer, tinystories_docs)
owt_tokens = tokenize_docs(owt_tokenizer, owt_docs)

# Tokenize cross-comparisons
tinystories_on_owt = tokenize_docs(tinystories_tokenizer, owt_docs)  # TinyStories tokenizer on OpenWebText
owt_on_tinystories = tokenize_docs(owt_tokenizer, tinystories_docs)  # OWT tokenizer on TinyStories

def compute_compression_ratio(docs, tokenized_docs):
    total_bytes = sum(len(doc.encode("utf-8")) for doc in docs)
    total_tokens = sum(len(tokens) for tokens in tokenized_docs)
    return total_bytes / total_tokens if total_tokens > 0 else float('inf')

# Compute compression ratios
tinystories_compression = compute_compression_ratio(tinystories_docs, tinystories_tokens)
owt_compression = compute_compression_ratio(owt_docs, owt_tokens)
cross_compression_tinystories = compute_compression_ratio(owt_docs, tinystories_on_owt)
cross_compression_owt = compute_compression_ratio(tinystories_docs, owt_on_tinystories)

# Print results
print(f"TinyStories Compression Ratio: {tinystories_compression:.2f} bytes/token")
print(f"OpenWebText Compression Ratio: {owt_compression:.2f} bytes/token")
print(f"Cross Compression (TinyStories on OWT): {cross_compression_tinystories:.2f} bytes/token")
print(f"Cross Compression (OWT on TinyStories): {cross_compression_owt:.2f} bytes/token")

import time

def estimate_tokenizer_throughput(tokenizer, docs):
    start_time = time.time()
    total_bytes = sum(len(doc.encode("utf-8")) for doc in docs)
    for doc in docs:
        _ = tokenizer.encode(doc)
    elapsed_time = time.time() - start_time
    return total_bytes / elapsed_time  # Bytes per second

# Measure tokenizer speeds
tinystories_throughput = estimate_tokenizer_throughput(tinystories_tokenizer, tinystories_docs)
owt_throughput = estimate_tokenizer_throughput(owt_tokenizer, owt_docs)

# Estimate time to tokenize 825GB
pile_size_bytes = 825 * 1024**3
tinystories_time = pile_size_bytes / tinystories_throughput / 3600  # Convert to hours
owt_time = pile_size_bytes / owt_throughput / 3600  # Convert to hours

# Print results
print(f"TinyStories Tokenizer Throughput: {tinystories_throughput:.2f} bytes/sec")
print(f"OpenWebText Tokenizer Throughput: {owt_throughput:.2f} bytes/sec")
print(f"Time to tokenize 825GB with TinyStories: {tinystories_time:.2f} hours")
print(f"Time to tokenize 825GB with OpenWebText: {owt_time:.2f} hours")

import numpy as np

def save_tokenized_data(tokenized_data, output_path):
    np.save(output_path, np.array(tokenized_data, dtype=object))  # dtype=object allows variable lengths
    print(f"Saved tokenized data to {output_path}")

# Save tokenized data
save_tokenized_data(tinystories_tokens, "tinystories_tokenized.npy")
save_tokenized_data(owt_tokens, "owt_tokenized.npy")
save_tokenized_data(tinystories_on_owt, "tinystories_on_owt.npy")
save_tokenized_data(owt_on_tinystories, "owt_on_tinystories.npy")
```

Script to calculate Parameters, Memory, and FLOPS of different gpt-2 variants:

```python
import numpy as np

def compute_params_and_flops(num_layers, d_model, num_heads, d_ff, vocab_size, context_length, tied_embedding=True):
    token_emb_params = vocab_size * d_model
    pos_emb_params = context_length * d_model
    attn_params = num_layers * 4 * d_model * d_model
    ffn_params = num_layers * 2 * d_model * d_ff
    norm_params = num_layers * 2 * d_model + d_model
    out_proj_params = 0 if tied_embedding else d_model * vocab_size

    total_params = token_emb_params + pos_emb_params + attn_params + ffn_params + norm_params + out_proj_params
    memory_gb = total_params * 4 / 1e9

    n, d_k = context_length, d_model / num_heads
    attn_proj_flops = num_layers * 8 * n * d_model * d_model / 1e12
    attn_dot_flops = num_layers * 4 * n * n * d_k * num_heads / 1e12
    ffn_flops = num_layers * 4 * n * d_model * d_ff / 1e12
    final_proj_flops = 2 * n * d_model * vocab_size / 1e12

    total_flops = attn_proj_flops + attn_dot_flops + ffn_flops + final_proj_flops

    return total_params, memory_gb, total_flops, attn_proj_flops, attn_dot_flops, ffn_flops, final_proj_flops

def print_summary(name, params, memory, total_flops, attn_proj, attn_dot, ffn, final_proj):
    print(f"\n{name}")
    print(f"Parameters: {params / 1e6:.2f}M")
    print(f"Memory (FP32): {memory:.2f} GB")
    print(f"Total Forward FLOPs: {total_flops:.2f} TFLOPs")
    print(f"  - Attention Projection FLOPs: {attn_proj:.2f} TFLOPs")
    print(f"  - Attention Dot-Product FLOPs: {attn_dot:.2f} TFLOPs")
    print(f"  - Feed-Forward FLOPs: {ffn:.2f} TFLOPs")
    print(f"  - Final Projection FLOPs: {final_proj:.2f} TFLOPs")

def main():
    models = {
        "GPT-2 Small": (12, 768, 12, 3072),
        "GPT-2 Medium": (24, 1024, 16, 4096),
        "GPT-2 Large": (36, 1280, 20, 5120),
        "GPT-2 XL": (48, 1600, 25, 6400),
    }
    vocab_size, context_length = 50257, 1024

    for name, (layers, d_model, heads, d_ff) in models.items():
        params, memory, total_flops, attn_proj, attn_dot, ffn, final_proj = compute_params_and_flops(layers, d_model, heads, d_ff, vocab_size, context_length)
        print_summary(name, params, memory, total_flops, attn_proj, attn_dot, ffn, final_proj)

    print("\nEffect of increasing context length to 16,384:")
    params, memory, total_flops, attn_proj, attn_dot, ffn, final_proj = compute_params_and_flops(48, 1600, 25, 6400, vocab_size, 16384)
    print_summary("GPT-2 XL (context=16,384)", params, memory, total_flops, attn_proj, attn_dot, ffn, final_proj)

if __name__ == "__main__":
    main()
```

```python
from collections.abc import Callable, Iterable
from typing import Optional
import torch
import math
import matplotlib.pyplot as plt

class SGD(torch.optim.Optimizer):
    def __init__(self, params, lr=1e-3):
        if lr < 0:
            raise ValueError(f"Invalid learning rate: {lr}")
        defaults = {"lr": lr}
        super().__init__(params, defaults)
    def step(self, closure: Optional[Callable] = None):
        loss = None if closure is None else closure()
        for group in self.param_groups:
            lr = group["lr"] # Get the learning rate.
            for p in group["params"]:
                if p.grad is None:
                    continue
                state = self.state[p] # Get state associated with p.
                t = state.get("t", 0) # Get iteration number from the state, or initial value.
                grad = p.grad.data # Get the gradient of loss with respect to p.
                p.data -= lr / math.sqrt(t + 1) * grad # Update weight tensor in-place.
                state["t"] = t + 1 # Increment iteration number.
            return loss


weights = torch.nn.Parameter(5 * torch.randn((10, 10)))
opt = SGD([weights], lr=10)
print("Learning weight of 10")
losses10 = []
for t in range(10):
    opt.zero_grad() # Reset the gradients for all learnable parameters.
    loss = (weights**2).mean() # Compute a scalar loss value.
    print(loss.cpu().item())
    losses10.append(loss.item())
    loss.backward() # Run backward pass, which computes gradients.
    opt.step() # Run optimizer step.
opt = SGD([weights], lr=100)
print("\nLearning weight of 100")
losses100 = []
for t in range(10):
    opt.zero_grad() # Reset the gradients for all learnable parameters.
    loss = (weights**2).mean() # Compute a scalar loss value.
    print(loss.cpu().item())
    losses100.append(loss.item())
    loss.backward() # Run backward pass, which computes gradients.
    opt.step() # Run optimizer step.
opt = SGD([weights], lr=1000)
print("\nLearning weight of 1000")
losses1000 = []
for t in range(10):
    opt.zero_grad() # Reset the gradients for all learnable parameters.
    loss = (weights**2).mean() # Compute a scalar loss value.
    print(loss.cpu().item())
    losses1000.append(loss.item())
    loss.backward() # Run backward pass, which computes gradients.
    opt.step() # Run optimizer step.

plt.figure(figsize=(6, 4))
plt.plot(losses10, label=f"LR=10")
plt.plot(losses100, label=f"LR=100")
plt.plot(losses1000, label=f"LR=1000")

plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.title("Loss vs. Iteration for Different Learning Rates")
plt.legend()
plt.tight_layout()
plt.show()
```
